```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Indian telecommunications market-1 - Sheet1.csv')
data.groupby('Location')['Support Government'].mean().plot(kind='bar')
plt.title('Rating for Supporting Government Locaion Wise')
plt.xlabel('Location')
plt.ylabel('Support Governtment')
plt.show()
```

```python
import matplotlib.pyplot as plt
data.groupby(['Location','Download Song']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Download Song vs Location')
plt.xlabel('Location')
plt.ylabel('Download Song')
plt.show()
```
